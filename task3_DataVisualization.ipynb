```python
import pandas as pd

data = {
    "Student": ["Asha", "Rahul", "Priya", "Arjun", "Meera"],
    "Marks": [85, 90, 78, 88, 92]
}

df = pd.DataFrame(data)

print(df)
```

```python
import matplotlib.pyplot as plt

plt.bar(df["Student"], df["Marks"])
plt.title("Student Marks")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.show()
```
