```python
import pandas as pd

data = {
    "Student": ["Asha", "Rahul", "Priya", "Arjun", "Meera"],
    "Marks": [85, 90, 78, 88, 92]
}

df = pd.DataFrame(data)

print(df)
```

```python
import matplotlib.pyplot as plt

plt.bar(df["Student"], df["Marks"])
plt.title("Student Marks")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.savefig("student_marks_bar_chart.png")
plt.show()
```

```python
import matplotlib.pyplot as plt

students = ["Asha", "Rahul", "Priya", "Arjun", "Meera"]
marks = [85, 90, 80, 89, 95]

plt.figure(figsize=(6,6))
plt.pie(marks, labels=students, autopct='%1.1f%%')
plt.title("Student Marks Distribution")
plt.savefig("student_marks_pie_chart.png")
plt.show()
```

```python
plt.figure(figsize=(7,5))
plt.plot(students, marks, marker='o')
plt.title("Student Marks Line Graph")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.savefig("student_marks_line_graph.png")
plt.show()
```
